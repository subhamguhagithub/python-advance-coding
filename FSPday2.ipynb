```python
class Worker:
  def __init__(self,name,hourly_rate):
     self.name=name
     self.hourly_rate=hourly_rate
     self.hours_worked=0 # Corrected typo from hours_workes to hours_worked
  def work(self,hours):
     self.hours_worked +=hours
     print(f"{self.name} worked {hours}.hours")
  def calculate_pay(self):
    return self.hourly_rate*self.hours_worked
```

```python
from enum import member
class Manager:
  def __init__(self,team_size=0):
      self.team_size=team_size
      self.team_members=[]
  def manage_team(self):
      print(f"Managing team of {self.team_size}people")
  def add_team_members(self,member):
      self.team_members.append(member)
      self.team_size=len(self.team_members)


```

```python
class TeamLead(Worker,Manager):
  def __init__(self,name,hourly_rate,team_size=0):
      Worker.__init__(self,name,hourly_rate)
      Manager.__init__(self,team_size)
      self.bonus=0
  def set_bonus(self, amount):
    self.bonus=amount
    print(f"{self.name} got a bonus of Inr{amount}")
  def calculate_pay(self):
      base_pay=Worker.calculate_pay(self)
      return base_pay+self.bonus

```

```python
lead=TeamLead("Anurag",100,10)
lead.work(40)
lead.add_team_members("Varun")
lead.add_team_members("Vivek")
lead.manage_team()
lead.set_bonus(10000)
print(f"total Pay:- Inr:{lead.calculate_pay()}")
```

Brute Force algorithm

```python
def Brute_force_search(number_list,target):
   for index in range(len(number_list)):
    print(f"checking index {index} : is {number_list[index]} equal to target {target}")
    #if number item matches the target we found it
    if number_list[index]==target:
      return index
    return -1
```

```python
my_lists=[5,2,9,7,1]
search_target=7
print(f"Searching Brute force search for {search_target} in the list: {my_lists}")
result_index=Brute_force_search(my_lists,search_target)
if result_index !=-1:
  print(f"Succes ! found {search_target} at index {result_index}")
else:
  print("Result not Found")
```

1. Find the largest element and the second largest  element in the array

```python
arr = [10, 20, 4, 45, 99]
largest = second = float("-inf")

for num in arr:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num

print(f"Largest element: {largest}")
print(f"Second largest element: {second}")
```

Reverse an Array

```python
arr=[1,2,3,4,5]
start=0
end=len(arr)-1
while(start<end):
  arr[start],arr[end]=arr[end],arr[start]# start=1, end=5=end=1,start=5
  start+=1
  end-=1
print(arr)
```

Check whether the given array is sorted in descending order

```python
arr = [5, 4, 3, 2, 1]

is_sorted = True

for i in range(1, len(arr)):
    if arr[i] > arr[i - 1]:
        is_sorted = False
        break

if is_sorted:
    print("The array is sorted in descending order")
else:
    print("The array is not sorted in descending order")
```

Check whether the given array is sorted in ascending order

```python
arr = [1, 2, 3, 4, 5]

is_sorted = True

for i in range(1, len(arr)):
    if arr[i] < arr[i - 1]:
        is_sorted = False
        break

if is_sorted:
    print("The array is sorted in ascending order")
else:
    print("The array is not sorted in ascending order")
```
