```python
class Student:

    """Docstring of the class"""

    def __init__(self):
        # instance variables
        self.name = "Anurag"
        self.roll_no = 101
        self.branch = "IT"

    def display(self):
        # instance method
        print("Student Details")
        print("The name is:", self.name)
        print("The roll number is:", self.roll_no)
        print("The branch is:", self.branch)


# object creation
s1 = Student()

# accessing instance variables
print(s1.name)
print(s1.roll_no)
print(s1.branch)

print()

# calling instance method
s1.display()
```

```python
class Value:
  def __init__(self):
    self.x=10
  def modify(self):
    self.x=self.x+1
  def display(self):
    print(self.x)
```

```python
v=Value()
v2=Value()
print(v.display()) #10
v.modify()
print(v.display()) #11
print(v2.display()) #10
```

```python
class Sample:
  x=10 # class variable or static variable
  @classmethod
  def modify(cls): #for creating class method
    cls.x=cls.x+1
```

```python
s1=Sample()
s2=Sample()
print(s1.x)
print(s2.x)
s1.modify()
print(s1.x)
print(s2.x)
```
