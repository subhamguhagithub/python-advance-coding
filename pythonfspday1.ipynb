```python
class Student:

    """Docstring of the class"""

    def __init__(self):
        # instance variables
        self.name = "Anurag"
        self.roll_no = 101
        self.branch = "IT"

    def display(self):
        # instance method
        print("Student Details")
        print("The name is:", self.name)
        print("The roll number is:", self.roll_no)
        print("The branch is:", self.branch)


# object creation
s1 = Student()

# accessing instance variables
print(s1.name)
print(s1.roll_no)
print(s1.branch)

print()

# calling instance method
s1.display()
```

```python
class Value:
  def __init__(self):
    self.x=10
  def modify(self):
    self.x=self.x+1
  def display(self):
    print(self.x)
```

```python
v=Value()
v2=Value()
print(v.display()) #10
v.modify()
print(v.display()) #11
print(v2.display()) #10
```

```python
class Sample:
  x=10 # class variable or static variable
  @classmethod
  def modify(cls): #for creating class method
    cls.x=cls.x+1
```

```python
s1=Sample()
s2=Sample()
print(s1.x)
print(s2.x)
s1.modify()
print(s1.x)
print(s2.x)
```

```python
### A python program using a student  class with instance method to  process data of several student(2)

# A python program using a student class
# with instance methods to process data of several students

class Student:

    def __init__(self, name, rollno, marks):
        self.name = name
        self.rollno = rollno
        self.marks = marks

    # method 1
    def display(self):
        print(f"My name is {self.name} and marks I got {self.marks}")

    # method 2
    def calculate(self):
        if self.marks >= 600:
            print("You got Grade A")
        elif self.marks > 500:
            print("You got Grade B")
        elif self.marks > 400:
            print("You got Grade C")
        else:
            print("You are fail")


```

```python
# taking input for several students
n = int(input("Enter the number of students: "))

i = 0
while(i < n):

    name = input("Enter the name of the student: ")
    rollno = input("Enter the roll number of the student: ")
    marks = int(input("Enter the marks of the student: "))

    # object creation
    s = Student(name, rollno, marks)

    # method calling
    s.display()
    s.calculate()

    print("---------------------------")

    i += 1
```

```python
#A python program to create a bank class where deposits and withdrawls can be handled by using instance method
import sys
class Bank:
  def __init__(self,name):
     self.name=name
     self.balance=0
  def deposit(self,amt):
    self.balance+=amt
    return self.balance
  def withdraw(self,amt):
    if amt>self.balance:
        return ("insufficient balance")
    self.balance-=amt
    return self.balance
```

```python
# create account
name = input("Enter the name: ")
b = Bank(name)

while True:
    print("\nd-deposit, w-withdraw, e-exit")
    choice = input("Enter your choice: ").lower()

    if choice == 'd':
        amt = float(input("Enter amount to deposit: "))
        print("Balance after deposit:", b.deposit(amt))

    elif choice == 'w':
        amt = float(input("Enter amount to withdraw: "))
        result = b.withdraw(amt)
        if isinstance(result, str):
            print(result)
        else:
            print("Balance after withdrawal:", result)

    elif choice == 'e':
        print("Thank you for using the bank.")
        break

    else:
        print("Invalid choice. Please try again.")
```

#An array given with value [1,2,2,3,1,1] how many same number is in the repeating form make sure the result store in the dictionary

```python

numbers = [1, 2, 2, 3, 1, 1]
counts = {}

for num in numbers:
    if num in counts:
        counts[num] += 1
    else:
        counts[num] = 1

print("Frequency dictionary:", counts)
```
