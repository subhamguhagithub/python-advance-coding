```python
### A python program using a student  class with instance method to  process data of several student(2)

# A python program using a student class
# with instance methods to process data of several students

class Student:

    def __init__(self, name, rollno, marks):
        self.name = name
        self.rollno = rollno
        self.marks = marks

    # method 1
    def display(self):
        print(f"My name is {self.name} and marks I got {self.marks}")

    # method 2
    def calculate(self):
        if self.marks >= 600:
            print("You got Grade A")
        elif self.marks > 500:
            print("You got Grade B")
        elif self.marks > 400:
            print("You got Grade C")
        else:
            print("You are fail")


```

```python
# taking input for several students
n = int(input("Enter the number of students: "))

i = 0
while(i < n):

    name = input("Enter the name of the student: ")
    rollno = input("Enter the roll number of the student: ")
    marks = int(input("Enter the marks of the student: "))

    # object creation
    s = Student(name, rollno, marks)

    # method calling
    s.display()
    s.calculate()

    print("---------------------------")

    i += 1
```

```python
#A python program to create a bank class where deposits and withdrawls can be handled by using instance method
import sys
class Bank:
  def __init__(self,name):
     self.name=name
     self.balance=0
  def deposit(self,amt):
    self.balance+=amt
    return self.balance
  def withdraw(self,amt):
    if amt>self.balance:
        return ("insufficient balance")
    self.balance-=amt
    return self.balance
```

```python
# create account
name = input("Enter the name: ")
b = Bank(name)

while True:
    print("\nd-deposit, w-withdraw, e-exit")
    choice = input("Enter your choice: ").lower()

    if choice == 'd':
        amt = float(input("Enter amount to deposit: "))
        print("Balance after deposit:", b.deposit(amt))

    elif choice == 'w':
        amt = float(input("Enter amount to withdraw: "))
        result = b.withdraw(amt)
        if isinstance(result, str):
            print(result)
        else:
            print("Balance after withdrawal:", result)

    elif choice == 'e':
        print("Thank you for using the bank.")
        break

    else:
        print("Invalid choice. Please try again.")
```

#An array given with value [1,2,2,3,1,1] how many same number is in the repeating form make sure the result store in the dictionary

```python

numbers = [1, 2, 2, 3, 1, 1]
counts = {}

for num in numbers:
    if num in counts:
        counts[num] += 1
    else:
        counts[num] = 1

print("Frequency dictionary:", counts)
```

```python
def bubble_sort(arr):
    length = len(arr)

    for i in range(0, length - 1):
        for j in range(0, length - i - 1):

            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr


if __name__ == "__main__":
    # Added .replace(',', ' ') to handle inputs like '1, 2, 3'
    user_input = input("Enter numbers separated by spaces or commas: ")
    arr = list(map(int, user_input.replace(',', ' ').split()))
    sort = bubble_sort(arr)

    print("Sorted array:", sort)
```

Binary Search (Divide and Conquer)

```python
def Binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = left + (right - left) // 2

        # target found
        if arr[mid] == target:
            return mid

        # search in right half
        elif arr[mid] < target:
            left = mid + 1

        # search in left half
        else:
            right = mid - 1

    # target not found
    return -1

# Example usage
numbers = [20, 40, 45, 55, 60, 62, 69, 72, 79]
print(f"Target 60 found at index: {Binary_search(numbers, 60)}")
```

Recursive method

```python
def binary_search_recursive(arr, target, left, right):
    if left > right:
        return -1

    mid = (left + right) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search_recursive(arr, target, mid + 1, right)
    else:
        return binary_search_recursive(arr, target, left, mid - 1)

# Example usage:
numbers = [20, 40, 45, 55, 60, 62, 69, 72, 79]
print(f"Target 72 found at index: {binary_search_recursive(numbers, 72, 0, len(numbers)-1)}")
```

```python
import random

def bubble_sort_optimized(n, heights):
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if heights[j] > heights[j + 1]:
                heights[j], heights[j + 1] = heights[j + 1], heights[j]
                swapped = True
        if not swapped:
            break
    return heights

def main():
    try:
        user_choice = input("Type 'auto' for a random list or 'manual' to type numbers: ").lower()

        if user_choice == 'auto':
            n = 20
            heights = [random.randint(1, 100) for _ in range(n)]
            print(f"Generated random list: {heights}")
        else:
            n = int(input("Enter the number of elements: "))
            print(f"Enter {n} numbers separated by spaces (do not use dots like '...'):")
            raw_input = input().replace(',', ' ')
            heights = [int(x) for x in raw_input.split()]
            n = len(heights)

        sorted_heights = bubble_sort_optimized(n, heights)
        print("Sorted result:", *(sorted_heights))

    except ValueError:
        print("Error: Invalid input. Please use only numbers and spaces (no dots or text).")

if __name__ == '__main__':
    main()
```

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

# Get user input and execute the sort
user_input = input("Enter numbers separated by spaces or commas: ")
# Convert input string to a list of integers
numbers = [int(x) for x in user_input.replace(',', ' ').split()]

sorted_list = selection_sort(numbers)
print("Sorted list using Selection Sort:", sorted_list)
```
