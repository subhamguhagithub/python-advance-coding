Inheritance

```python
class Parent:
   def show(self):
       print("This is parent class")
class Child(Parent):
   def display(self):
       print("This is child class")
```

```python
c=Child()
c.show()
c.display()
```

```python
class Worker:
    def __init__(self, name, hourly_rate):
        self.name = name
        self.hourly_rate = hourly_rate
        self.hours_worked = 0

    def work(self, hour):
        self.hours_worked += hour

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

    def display(self):
        print("Name:", self.name)
        print("Hourly Rate:", self.hourly_rate)
        print("Hours Worked:", self.hours_worked)
        print("Salary:", self.calculate_salary())


# Driver Code
w = Worker("Shubham", 200)

w.work(5)
w.work(3)

w.display()
```

Python program to replace every nth index position charecter

```python
s = input("Enter string: ")
x = input("Enter replacement character: ")
n = int(input("Enter n: "))

result = ""

for i in range(len(s)):
    if i % n == 0 and i != 0:
        result += x
    else:
        result += s[i]

print(result)
```

```python
def replace_nth_char(string, char, n):
    """Replace every nth character in a string with replacement char"""

    if n <= 0:
        return "Error: n must be greater than 0"

    result = ""  # empty string

    for i in range(len(string)):
        if (i + 1) % n == 0:   # every nth character (1-based)
            result += char
        else:
            result += string[i]

    return result


# Example
s = "welcome to python coding"
print(replace_nth_char(s, "*", 3))
```
